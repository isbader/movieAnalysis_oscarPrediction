```python
#Plot a heat map of the missing varibles
def missing_heat_map(DataFrame):
    fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (18, 6))
    sns.heatmap(DataFrame.isnull(), yticklabels=False, ax = ax, cbar=False, cmap='viridis')
    ax.set_title('Movies_')
    plt.show()
    
    for i in DataFrame:
        print(i,': %',int((DataFrame[i].isnull().sum()/len(DataFrame[i]))*100),\
              'With {} missing values'.format((DataFrame[i].isnull().sum())))
missing_heat_map(Movies_)
```

```python
import ast

imd = [] #stands for Internet Movie Database
rotten = [] #rotten tometo
metacritic = []

for i in df['Ratings']:
    if type(i) == str:
        a = ast.literal_eval(i)
            
        if len(a) > 2:
            imd.append(a[0])
            rotten.append(a[1])
            metacritic.append(a[2])
        elif len(a) > 1:
            imd.append(a[0])
            metacritic.append(a[1])
            rotten.append(None)
        elif len(a) == 1:
            imd.append(a[0])
            metacritic.append(None)
            rotten.append(None)    
        else:
            imd.append(None)
            rotten.append(None)
            metacritic.append(None)
    else:
        imd.append(None)
        rotten.append(None)
        metacritic.append(None)
        
```

## Replace actor's names with awards count 

```python
def names_to_count(df,col,win,nom):
    if 'Unnamed: 0' in nom.columns:
        del nom['Unnamed: 0']
    if 'Unnamed: 0' in win.columns:
        del win['Unnamed: 0'] 
    if 'id' in nom.columns:
        del nom['id']
    if 'id' in win.columns:
        del win['id']
    
        
    win.columns = (['2000','2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
           '2014', '2015', '2016', '2017', '2018', '2019', 'i'])
    nom.columns =  (['2000','2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
           '2014', '2015', '2016', '2017', '2018', '2019',  'i'])

    win.set_index('i')
    nom.set_index('i')

    tempact = pd.DataFrame()
    tempact['per'] = df[col]
    perwin = []
    pernom = []
    perw = []
    y = []
    name = []

    for index, row in df.iterrows():
        this_movie = str(row['fname'])
        this_year = str(row['fyear'])
        y.append(this_year)
        name.append(this_movie)
        try:
            w = list(win.loc[win['i'] == row[col]][this_year])[0]
            perwin.append(w)
        except:
            perwin.append(0)
        try:
            n = list(nom.loc[win['i'] == row[col]][this_year])[0]
            pernom.append(n)
        except:
            pernom.append(0)
        perw.append(perwin[-1] + pernom[-1]*.5)

    
    tempact['mname'] = name
    tempact['year'] = y
    tempact[col+'win'] = perwin
    tempact[col +'nom'] = pernom
    tempact[col +'w'] = perw
    return perw

```
