```python
def create_matrix(df,name):
    names = df.loc[0]
    
    win = df.loc[1:]
    nom = df.loc[1:]
    
    winner = win.transpose()
    nominate = nom.transpose()
    
    win_col =  winner.columns
    nom_col =  nominate.columns
    
    for w in win_col:
        winner[w] = winner[w].apply(lambda col : col.count('Winner') if type(col) == str else 0)
        
    for n in nom_col:
        nominate[n] = nominate[n].apply(lambda col : col.count('Nominee') if type(col) == str else 0)
        
    winner['id']= winner.index
    nominate['id'] = nominate.index
    wn = list( df.loc[0])
    winner['i']= wn
    nominate['i']= wn
    
    winner.set_index('i')
    nominate.set_index('i')
    
    nominate.drop_duplicates(keep='first',inplace=True)
    winner.drop_duplicates(keep='first',inplace=True)
    
    return winner.to_csv('{}_new_winner.csv'.format(name),index=True,header=True),nominate.to_csv('{}_new_nominate.csv'.format(name),index=True,header=True)


    
    
```
