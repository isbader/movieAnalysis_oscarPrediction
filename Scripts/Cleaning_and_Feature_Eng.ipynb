```python
import numpy as np
import pandas as pd
import imdb
```

```python
Movies = pd.read_csv('Movies_2006_castfixed.csv') #Get the movies
```

### Language columns

```python
Movies.head(1)
```

```python
Movies['en'] =  Movies[Movies['language_codes'].notnull()]['language_codes'].apply(lambda col: 1 if 'en' in col else 0)
```

```python
Movies[['language_codes','en']].head()
```

```python
Movies['multi_lang'] =  Movies[Movies['language_codes'].notnull()]['language_codes'].apply(lambda col: len(col.split(',')))
```

```python
Movies[['language_codes','multi_lang']].head()
```

### Country columns

```python
Movies['US'] =  Movies[Movies['country_codes'].notnull()]['country_codes'].apply(lambda col: 1 if 'us' in col else 0)
```

```python
Movies[['country_codes','US']].head()
```

```python
Movies['Other_countries'] =  Movies[Movies['country_codes'].notnull()]['country_codes'].apply(lambda col: 1 if 'us' in col else 0)
```

### Split OWUS column to month and OWUS

```python
Movies['OWUS'].head(1)
```

Remove dollar sign and split the date  from OWUS:

```python

Movies['month'] =  Movies[Movies['OWUS'].notnull()]['OWUS'].apply(lambda col : col.split(',')[-1].split(' ')[2] )
Movies['OWUS'] = Movies[Movies['OWUS'].notnull()]['OWUS'].apply(lambda col : col.replace(',','').replace('$','').replace(col.split(',')[-1],''))
```

```python
Movies['OWUS'].head()
```

### Add Oscar/nominated column 

```python
OscarList =  pd.read_csv('../RelatedWork/TheOscarAward/the_oscar_award.csv') #Get the oscar list
```

```python
ll = OscarList[8400:]
ll.columns
```

```python
MoviesOscar = pd.merge(left=Movies,right=ll, how='left', left_on='name', right_on='film')
MoviesOscar.head(1)
```

```python
del MoviesOscar['year_ceremony']
del MoviesOscar['ceremony']
del MoviesOscar['category']
del MoviesOscar['name_y']
del MoviesOscar['year_film']    
```

```python
MoviesOscar.head()
```

```python
MoviesOscar['film'].loc[~MoviesOscar['film'].isnull()] = 1  # not nan
MoviesOscar['film'].loc[MoviesOscar['film'].isnull()] = 0   # nan
MoviesOscar['win'].loc[MoviesOscar['win'].isnull()] = 0   # nan

dd = {True: 1, False: 0}
MoviesOscar["win"]= MoviesOscar["win"].map(dd) 
```
